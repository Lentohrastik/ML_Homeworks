Задание 14

```python
df['Math_PassStatus'] = df['math score'].apply(lambda score: 'F' if score < passmark else 'P')
df['Reading_PassStatus'] = df['reading score'].apply(lambda score: 'F' if score < passmark else 'P')
df['Writing_PassStatus'] = df['writing score'].apply(lambda score: 'F' if score < passmark else 'P')
```

```python
math_pass = df[df['Math_PassStatus'] == 'P'].shape[0]
reading_pass = df[df['Reading_PassStatus'] == 'P'].shape[0]
writing_pass = df[df['Writing_PassStatus'] == 'P'].shape[0]
print('Количество студентов, сдавших/не сдавших экзамен по математике:', math_pass, '/', df.shape[0] - math_pass)
print('Количество студентов, сдавших/не сдавших экзамен по математике:', reading_pass, '/', df.shape[0] - reading_pass)
print('Количество студентов, сдавших/не сдавших экзамен по математике:', writing_pass, '/', df.shape[0] - writing_pass)
```

Задание 15

```python
def func(data):
    if data['Math_PassStatus'] == 'P' and  data['Reading_PassStatus'] == 'P' and  data['Writing_PassStatus'] == 'P':
        return 'P'
    return 'F'
df['OverAll_PassStatus'] = df.apply(func, axis=1)
```

```python
print('Количество студентов, которые сдали все экзамены:', df['OverAll_PassStatus'].map({'P': 1, 'F' : 0}).sum())
```

Задание 16

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    elif average_mark >= 80:
        return 'B'
    elif average_mark >= 70:
        return 'C'
    elif average_mark >= 60:
        return 'D'
    elif average_mark >= 50:
        return 'E'
    return 'F'

df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
df
```
